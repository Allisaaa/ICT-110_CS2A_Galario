EXERCISE 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    


```python
hours = float(input('Enter Hours: '))
rate = float(input('Enter Rate: '))

if hours <= 40:
      pay = hours * rate

else:
      regular_pay = 40 * rate
      overtime_pay = (hours - 40) * (1.5 * rate)
      pay = regular_pay + overtime_pay

print('Pay: ', pay)
```

CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0




```python
try:
  score = float(input('Enter score (between 0 and 100): '))

  if 97 <= score < 100 :
      grade = 1.0
  elif 93 <= score < 97 :
      grade = 1.25
  elif 89 <= score < 93 :
      grade = 1.5
  elif 85 <= score < 89 :
      grade = 1.75
  elif 81 <= score < 85 :
      grade = 2.0
  elif 77 <= score < 81 :
      grade = 2.25
  elif 74 <= score < 77 :
      grade = 2.5
  elif 0 <= score < 74 :
      grade = 3.0
  else:
    print('Invalid score entered.')
    exit()

  print("Grade: ", grade)

except ValueError:
  print('Please enter a valid score:')
```
