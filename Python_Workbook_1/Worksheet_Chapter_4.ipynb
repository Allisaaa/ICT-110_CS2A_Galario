EXERCISE 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    


```python
hours = float(input('Enter Hours: '))
rate = float(input('Enter Rate: '))

def compute_pay(hours, rate):

    if hours <= 40:
        pay = rate * hours

    else:
        reg_pay = 40 * rate
        overtime_pay = (hours - 40) * (1.5 * rate)
        pay = reg_pay + overtime_pay

    return pay

pay = compute_pay(hours, rate)
print('Pay: ', pay)

```

CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    

<br>

```python
score = float(input('Enter score (between 0 and 100): '))

def computegrade(score):
    if 97 <= score <= 100:
        grade = '1.0'
    elif 93 <= score < 97:
        grade = '1.25'
    elif 89 <= score < 93:
        grade = '1.5'
    elif 85 <= score < 89:
        grade = '1.75'
    elif 81 <= score < 85:
        grade = '2.0'
    elif 77 <= score < 81:
        grade = '2.25'
    elif 74 <= score < 77:
        grade = '2.5'
    else:
        grade = '3.0'
    return grade

result = computegrade(score)
print('Grade:', result)
```
